```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
number_of_schools=school_data_complete["school_name"].nunique()
number_of_students = school_data_complete["student_name"].count()
total_budget = school_data["budget"].sum()
average_math_score = school_data_complete["math_score"].mean()
average_reading_score = school_data_complete["reading_score"].mean()




summary_dict = [{"Number of Schools":number_of_schools,
                 "Number of Students":number_of_students,
                 "Total Budget":total_budget,
                 "Average Math Score": average_math_score,
                 "Average Readiing Score": average_reading_score,
                
                }]

pd.DataFrame(summary_dict)
#top_30_math_score= school_data_complete.sort_values("math_score", ascending = False)
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
a = school_data_complete.groupby(['school_name']).first()

reading_score_average = a['reading_score'].mean()
reading_score_average

math_score_average = a['math_score'].mean()
math_score_average

number_of_students

student_budget = a['budget'].mean()/a['size'].mean()
student_budget

total_budget = number_of_students * student_budget
total_budget

a1 = school_data_complete.loc[school_data_complete['math_score'] >=70].groupby(['school_name']).size()
percenatge_passed_in_maths = a1/number_of_students*100

a2 = school_data_complete.loc[school_data_complete['reading_score'] >=70].groupby(['school_name']).size()
percenatge_passed_in_reading = a2/number_of_students*100

a3 = school_data_complete.loc[(school_data_complete['math_score'] >=70) & (school_data_complete['reading_score'] >=70)].groupby(['school_name']).size()
percenatge_passed_overall = a3/number_of_students*100

summary = pd.DataFrame({'School Type': a['type'],
                        'Total Students': number_of_students,
                        'Total Budget' : total_budget,
                       'Per Student Budget' : student_budget,
                       'Average Math Score' : math_score_average,
                       'Average Reading Score' : reading_score_average,
                        '% Passing Math' : percenatge_passed_in_maths,
                        '% Passing Reading' : percenatge_passed_in_reading,
                        '% Overall Passing' : percenatge_passed_overall,
                       })

copy_of_summary = summary.copy()
copy_of_summary['Total Budget'] = copy_of_summary['Total Budget'].map("${:,.2f}".format)
copy_of_summary['Per Student Budget'] = copy_of_summary['Per Student Budget'].map("${:,.2f}".format)

copy_of_summary['Average Math Score'] = copy_of_summary['Average Math Score'].map("{:,.2f}".format)
copy_of_summary['Average Reading Score'] = copy_of_summary['Average Reading Score'].map("{:,.2f}".format)
copy_of_summary['% Passing Math'] = copy_of_summary['% Passing Math'].map("{:,.2f}%".format)
copy_of_summary['% Passing Reading'] = copy_of_summary['% Passing Reading'].map("{:,.2f}%".format)
copy_of_summary['% Overall Passing'] = copy_of_summary['% Overall Passing'].map("{:,.2f}%".format)

copy_of_summary

```
